## **Regression Models**

### **OLS Model**

We will construct an Ordinary Least Squares (OLS) model using the Popular percentage as the dependent variable. Once the model is built, we will display and analyze the regression results to evaluate its performance and interpret the relationships between the independent variables and the electoral outcomes.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np


X = df.iloc[:, 3:]


Y = df.iloc[:, 2]


X = sm.add_constant(X)


```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
model = sm.OLS(Y,X).fit()
print(model.summary())
```

To assess the performance of the **OLS model**, we will plot the predicted values of the popular vote percentage against the actual values. This visualization will help us evaluate how well the model's predictions align with the observed data. We will calculate the Root Mean Squared Error *(RMSE)* and Mean Absolute Error *(MAE)* to quantitatively assess the model's performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression

y_pred = model.predict(X)
rmse = np.sqrt(mean_squared_error(Y, y_pred))
print(f"RMSE: {rmse}")

mae = mean_absolute_error(Y, y_pred)
print(f"MAE: {mae}")

plt.plot(Y, label='Actual Y', color='blue')
plt.plot(y_pred, label='Predicted Y', color='red')
plt.xlabel('election cycles')
plt.ylabel('popular vote percentage of incumbent party ')
plt.title('Actual vs Predicted Values of popular percentage')
plt.legend()
plt.show()
```

### **LASSO Model**

We will construct a **LASSO** (Least Absolute Shrinkage and Selection Operator) model to improve upon the results of our OLS model. After developing the model, we will plot predicted values against actual ones to assess performance. We will also use the model to predict new data.

```python
import pandas as pd
from sklearn.linear_model import Lasso, LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler


lasso_cv = LassoCV(cv=5, random_state=41)
lasso_cv.fit(X, Y)
optimal_lambda = lasso_cv.alpha_
print(f'Optimal λ: {optimal_lambda}')


lasso_model = Lasso(alpha=0.0002)
lasso_model.fit(X, Y)
y_pred_lasso = lasso_model.predict(X)

feature_names = [f'Feature_{i}' for i in range(X.shape[1])]

lasso_coefficients = lasso_model.coef_

coefficients_df = pd.DataFrame({
    'Variable': feature_names,
    'Coefficient': lasso_coefficients
})
print(coefficients_df)

rmse_lasso = np.sqrt(mean_squared_error(Y, y_pred_lasso))
mae_lasso = mean_absolute_error(Y, y_pred_lasso)
r2_lasso = r2_score(Y, y_pred_lasso)
r_adj = 1 - (1-lasso_model.score(X, Y))*(len(Y)-1)/(len(Y)-X.shape[1]-1)


print(f"RMSE: {rmse_lasso}")
print(f"MAE: {mae_lasso}")
print(f"R-squared: {r2_lasso}")
print(f"adjusted R-squared: {r_adj}")

plt.plot(Y, label='Actual Y', color='blue')
plt.plot(y_pred_lasso, label='Predicted Y', color='red')
plt.xlabel('Index')
plt.ylabel('Value of Y')
plt.title('Actual vs. Predicted Values (OLS Regression)')
plt.legend()
plt.show()


excel_file_path = '/content/Book pred 2.xlsx'
sheet_name = 'Sheet1'
dfpred = pd.read_excel(excel_file_path, sheet_name=sheet_name)
xpred = dfpred.iloc[:, 3:]
xpred= sm.add_constant(xpred)

pred = lasso_model.predict(xpred)
pred
```

## **Classification Models**

To develop classification models, we converted the dependent variable to binary: 1 for a percentage exceeding 50% (win) and 0 otherwise. We built *ANN*, *Naive Bayes*, *Logistic Regression*, and *Decision Tree* models, evaluated their performance using metrics like accuracy, precision, and F1 score, and used them to predict new data.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


split_index = int(0.6 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = Y[:split_index], Y[split_index:]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='binary')
    f1 = f1_score(y_test, y_pred, average='binary')
    return accuracy, precision, f1

ann_model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

ann_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann_model.fit(X_train_scaled, y_train, epochs=100, verbose=0)

y_pred_ann = (ann_model.predict(X_test_scaled) > 0.5).astype("int32")

accuracy_ann = accuracy_score(y_test, y_pred_ann)
precision_ann = precision_score(y_test, y_pred_ann, average='binary')
f1_ann = f1_score(y_test, y_pred_ann, average='binary')

nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)
y_pred_nb = nb_model.predict(X_test_scaled)

accuracy_nb, precision_nb, f1_nb = evaluate_model(nb_model, X_test_scaled, y_test)

lr_model = LogisticRegression()
lr_model.fit(X_train_scaled, y_train)
y_pred_lr = lr_model.predict(X_test_scaled)

accuracy_lr, precision_lr, f1_lr = evaluate_model(lr_model, X_test_scaled, y_test)

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test_scaled)

accuracy_dt, precision_dt, f1_dt = evaluate_model(dt_model, X_test, y_test)

print(f"ANN - Accuracy: {accuracy_ann}, Precision: {precision_ann}, F1 Score: {f1_ann}")
print(f"Naive Bayes - Accuracy: {accuracy_nb}, Precision: {precision_nb}, F1 Score: {f1_nb}")
print(f"Logistic Regression - Accuracy: {accuracy_lr}, Precision: {precision_lr}, F1 Score: {f1_lr}")
print(f"Decision Tree - Accuracy: {accuracy_dt}, Precision: {precision_dt}, F1 Score: {f1_dt}")

print("\nComparison of Predictions to Actual Test Values:\n")
print("True Values:\n", y_test)
print("\nANN Predictions:\n", y_pred_ann.flatten())
print("\nNaive Bayes Predictions:\n", y_pred_nb.flatten())
print("\nLogistic Regression Predictions:\n", y_pred_lr.flatten())
print("\nDecision Tree Predictions:\n", y_pred_dt.flatten())


excel_file_path = '/content/Book2 test.xlsx'
sheet_name = 'Sheet1'
dfpred = pd.read_excel(excel_file_path, sheet_name=sheet_name)
xpred = dfpred.iloc[:, 3:]
xpred= sm.add_constant(xpred)

new_data_scaled = scaler.transform(xpred)

prediction = (ann_model.predict(new_data_scaled) > 0.5).astype("int32")

print("Predicted outcome:", prediction)

nb_predicted_outcome = nb_model.predict(new_data_scaled)

print("Predicted outcome:", nb_predicted_outcome)

predicted_outcome = lr_model.predict(new_data_scaled)

print("Predicted outcome:", predicted_outcome)

predicted_outcome = dt_model.predict(new_data_scaled)

print("Predicted outcome:", predicted_outcome)
```
